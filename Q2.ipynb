```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([0,0])
Y = np.array([0,0])
U = np.array([1,2])
V = np.array([3,-1])

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.axhline(0, color='blue', linewidth=1)
plt.axvline(0, color='blue', linewidth=1)

plt.xticks([-5, -4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-5, -4, -3, -2, -1, 0, 1, 2, 3, 4])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Span of 2 vectors')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.grid()
plt.show()
```

در این بخش ابتدا لایبری های مورد نظر را ایمپورت کرده و سپس به تعریف مبدا مختصات و بردار های ابتدایی که در سوال آمده پرداخته ایم و در نهایت با استفاده از تابع ذکر شده توانستیم این ها را به صورت بصری نشان دهیم.

```python

scaler = np.arange(-5,5,0.001)
scaler = scaler.reshape(scaler.shape[0],1)
np.random.shuffle(scaler)

scaler2 = np.arange(-5,5,0.001)
scaler2 = scaler2.reshape(scaler2.shape[0],1)
np.random.shuffle(scaler2)

span_x = U[0]*scaler + U[1]*scaler2
span_y = V[0]*scaler + V[1]*scaler2
```

در این بخش هم ابتدا یک سری ضریب برای ضرب کردن درون بردار های ابتدایی تولید میکنیم که در آن رنج را مشخص میکنیم از چه عدد تا چه عدد باشد را مشخص می‌کنیم . با کوچک تر کردن آرگومان سوم تابع ارنج مقدار اختلاف بین دو نقطه کاهش می‌یابد و نقاط تولیدی بیشتری تولید می‌شود در نتیجه اسپن به شکل گسترده تری نمایش داده می‌شود. در واقع استپ های بین دو عدد را مشخص میکند خروجی های برنامه به ازای مقادیر مختلف اسکرین شات گرفته شده و در فولدر زیپ شده قرار داده شده است

```python

plt.scatter(span_x,span_y, color='b', label='data points' , marker= 'o')
plt.legend()

plt.axhline(0, color='blue', linewidth=1)
plt.axvline(0, color='blue', linewidth=1)

plt.xticks([-5, -4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-5, -4, -3, -2, -1, 0, 1, 2, 3, 4])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Span of 2 vectors')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.grid()
plt.show()

```
